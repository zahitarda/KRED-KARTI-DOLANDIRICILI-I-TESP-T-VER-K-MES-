#### KREDİ KARTI DOLANDIRICILIĞI TESPİT VERİ KÜMESİ

```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("credit_card_fraud.csv")

df = df.rename(columns={
    "amount": "miktar",
    "transaction_hour": "işlem_saati",
    "foreign_transaction": "yabancı_işlem",
    "location_mismatch": "konum_uyumsuzluğu",
    "device_trust_score": "cihaz_güven_puanı",
    "velocity_last_24h": "işlem_sayısı(son24sa)",
    "cardholder_age": "kart_sahibinin_yaşı",
    "is_fraud": "dolandırıcılık"
})

print("Veri setinin ilk 5 satırı:")

print(df.head())
print()
print("\n sütunlar:", df.columns.tolist())
```

```python
X = df[["miktar", "işlem_saati", "yabancı_işlem", "konum_uyumsuzluğu", "cihaz_güven_puanı", "işlem_sayısı(son24sa)"]]

y = df["dolandırıcılık"]

print(X.head())
print()
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,  
    y,  
    test_size = 0.2,  
    random_state = 25
)

print(X_train.shape, y_train.shape)
print()
print(X_test.shape, y_test.shape)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)

knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)

print(y_pred[:750])
print()
print(y_test.values[:750])
```

```python
acc = accuracy_score(y_test, y_pred)
print("Model doğruluğu:", acc)

sonuçlar = X_test.copy()
sonuçlar["Gerçek (pass_fail)"] = y_test.values
sonuçlar["Tahmin (pass_fail)"] = y_pred

print("Test verisinden örnek karşılaştırma (ilk 10 satır).")
print(sonuçlar.head(750))
```
